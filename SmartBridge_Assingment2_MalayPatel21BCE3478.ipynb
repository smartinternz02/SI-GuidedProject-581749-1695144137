```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
file_path = "/content/House Price India.csv"
df = pd.read_csv(file_path)
```

```python
# Univariate Analysis
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'])
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Box Plot (Box-and-Whisker Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Price'])
plt.title('Box Plot of Price')
plt.xlabel('Price')
plt.show()

# Kernel Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Price'], shade=True)
plt.title('Kernel Density Plot of Price')
plt.xlabel('Price')
plt.show()
```

```python
# Bivariate Analysis
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['lot area'], y=df['Price'])
plt.title('Scatter Plot of Area vs. Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()

# Heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Pairplot
sns.pairplot(df[['lot area', 'Price', 'number of bedrooms', 'number of bathrooms']])
plt.title('Pairplot')
plt.show()
```

```python
# Multivariate Analysis
# Parallel Coordinates Plot
fig = px.parallel_coordinates(df, dimensions=['lot area', 'Price', 'number of bedrooms', 'number of bathrooms'], color="Price")
fig.show()

# 3D Scatter Plot
fig = px.scatter_3d(df, x='lot area', y='number of bedrooms', z='Price', color='Price')
fig.show()

# Radial Plot (Polar plot)
fig = px.line_polar(df, r='Price', theta='lot area', line_close=True)
fig.update_traces(fill='toself')
fig.show()
```
